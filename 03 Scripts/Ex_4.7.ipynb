### 2. user defined functions

```python
#conditions for user defined functions
#If the item’s price is lower than or equal to $5, it will be labeled a “low-range product.”
#If the item’s price is above $5 but lower than or equal to $15, it will be labeled a “mid-range product.”
#If the item’s price is above $15, it will be labeled a “high-range product.”
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# using the function on the 1000000 dataframe
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

### 5. exercise deriving variables

#### 5.1 changing the labels 

```python
# Create 'busiest_days' column instead of 'busiest_day' from 'orders_day_of_week' as follows, days 0 and 1 as busiest, days 3 and 4 as slowest days. 
#The remainder of the days as regularly busy
```

```python
#Using loops to make the changes
```

```python
result__ = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result__.append("Busiest days")
  elif value == 3 or value == 4:
    result__.append("Slowest days")
  else:
    result__.append("Regularly busy")
```

```python
result__
```

```python
#checking if the column has the new day classifications
```

```python
#Rename the column 'busiest_day' to 'busiest_days' in df_ords_prods_merge
```

```python
df_ords_prods_merge['busiest_days'] = result__
```

```python
df_ords_prods_merge.head()
```

```python
#Question 3. Check the values of this new column for accuracy. Note any observations in markdown format.
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

The major realization is that, each classification of "busiest_days" is now a sum of orders for days under "Busiest days", "Slowest days" and "Regularly busy". For example, "Busiest days" has value counts of 11 875 462 which is a sum of day 0 (6 209 632) and day 1 (5 665 830). The same is true for days 3 and 4 (Slowest days).

```python
#Question 4. identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” 
#Create a new column containing these labels called “busiest_period_of_day.”
```

```python
# Check distribution of orders by hour of the dayr "order_hour_of_day" column.

df_ords_prods_merge['order_hour_of_day'].value_counts()
```

```python
#The day has 24 hours and the new classification should have 6 hours each.Most orders should have 10, 11, 14, 15,13, 12; 
#Fewest Order has 3,4,2,5,1,0 and the remaining hours are for Avergage orders 
```

```python
result_2 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_2.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_2.append("Fewest orders")
  else:
    result_2.append("Average orders")
```

```python
result_2
```

```python
# Creating a new column "busiest_period_of_day" in df_ords_prods_merge dataframe.

df_ords_prods_merge['busiest_period_of_day'] = result_2
```
