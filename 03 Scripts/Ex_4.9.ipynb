### 3. creating histograms and scatterplots

```python
# creating a histogram of the prices column
ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
ords_prods_merge['prices'].describe()
```

```python
ords_prods_merge['prices'].mean()
```

```python
ords_prods_merge['prices'].median()
```

```python
ords_prods_merge['prices'].max()
```

extreme value of 99,999 is forcing your histogram to shove all the small values into one bar and the extreme value in another one.

```python
#doing a scatter plot to see distribution of prices
```

```python
sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_merge)
```

```python
#checking for datapoints above 100
ords_prods_merge.loc[ords_prods_merge['prices'] > 100]
```

```python
# turning all outliers as NaNs
ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
ords_prods_merge['prices'].max()
```

```python
# creating a histogram after relegating outliers
ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
# increasing granulity of histogram by adding more bins
ords_prods_merge['prices'].plot.hist(bins = 70)
```

```python
#export the histogram as an image file
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_prices.png'))
```
